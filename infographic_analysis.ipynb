```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df_accidents_injuries_2010 = pd.read_excel('./data/road_accidents_injury_severity.xlsx', sheet_name='2010')
numeric_columns = df_accidents_injuries_2010.select_dtypes(include=[np.number]).columns.tolist()
for column in numeric_columns:
    df_accidents_injuries_2010[column] = df_accidents_injuries_2010[column].astype(int)

df_accidents_injuries_2011 = pd.read_excel('./data/road_accidents_injury_severity.xlsx', sheet_name='2011')
numeric_columns = df_accidents_injuries_2011.select_dtypes(include=[np.number]).columns.tolist()
for column in numeric_columns:
    df_accidents_injuries_2011[column] = df_accidents_injuries_2011[column].astype(int)


df_accidents_injuries_2012 = pd.read_excel('./data/road_accidents_injury_severity.xlsx', sheet_name='2012')
numeric_columns = df_accidents_injuries_2012.select_dtypes(include=[np.number]).columns.tolist()
for column in numeric_columns:
    df_accidents_injuries_2012[column] = df_accidents_injuries_2012[column].astype(int)

df_accidents_vehicles_2010 = pd.read_excel('./data/road_accidents_vehicle_type.xlsx', sheet_name='2010')
numeric_columns = df_accidents_vehicles_2010.select_dtypes(include=[np.number]).columns.tolist()
for column in numeric_columns:
    df_accidents_vehicles_2010[column] = df_accidents_vehicles_2010[column].astype(int)
    
df_accidents_vehicles_2011 = pd.read_excel('./data/road_accidents_vehicle_type.xlsx', sheet_name='2011')
numeric_columns = df_accidents_vehicles_2011.select_dtypes(include=[np.number]).columns.tolist()
for column in numeric_columns:
    df_accidents_vehicles_2011[column] = df_accidents_vehicles_2011[column].astype(int)

df_accidents_vehicles_2012 = pd.read_excel('./data/road_accidents_vehicle_type.xlsx', sheet_name='2012')
numeric_columns = df_accidents_vehicles_2012.select_dtypes(include=[np.number]).columns.tolist()
for column in numeric_columns:
    df_accidents_vehicles_2012[column] = df_accidents_vehicles_2010[column].astype(int)

df_accidents_injuries_2012.head(30)
df_accidents_vehicles_2012.head(30)
```

```python
df_accidents_injuries_2010['Total_deaths'] = df_accidents_injuries_2010['Deaths Male'] + df_accidents_injuries_2010['Deaths Female'] 
df_accidents_injuries_2010.head(30)

df_accidents_injuries_2012['Total_deaths'] = df_accidents_injuries_2012['Deaths Male'] + df_accidents_injuries_2012['Deaths Female'] 
df_accidents_injuries_2012.head(30)
```

```python
#merge the df_accidents_injuries data frames on Districts and have the Total columns separate for 2010, 2011 and 2012
df_accidents_injuries_2010_2011 = pd.merge(df_accidents_injuries_2010, df_accidents_injuries_2011, on='Districts', suffixes=('_2010', '_2011'))
df_accidents_injuries_2010_2011.head(30)
#add _2012 suffix to the columns of df_accidents_injuries_2012
df_accidents_injuries_2012.columns = [str(col) + '_2012' for col in df_accidents_injuries_2012.columns]
df_accidents_injuries_2012.head(30)
#rename the Districts_2012 column to Districts
df_accidents_injuries_2012.rename(columns={'Districts_2012':'Districts'}, inplace=True)
df_accidents_injuries_2010_2011_2012 = pd.merge(df_accidents_injuries_2010_2011, df_accidents_injuries_2012, on='Districts') 
df_accidents_injuries_2010_2011_2012.head(30)
#create new df with the total columns for 2010, 2011 and 2012 and Districts
df_accidents_injuries_2010_2011_2012_total = df_accidents_injuries_2010_2011_2012[['Districts', 'Total_2010', 'Total_2011', 'Total_2012']]
#rename Total columns to Total_accidents_2010, Total_accidents_2011, Total_accidents_2012
df_accidents_injuries_2010_2011_2012_total.rename(columns={'Total_2010':'Total_injuries_2010', 'Total_2011':'Total_injuries_2011', 'Total_2012':'Total_injuries_2012'}, inplace=True)
df_accidents_injuries_2010_2011_2012_total.head(30)
```

```python
#merge the vehicles data frames on Districts and have the Total columns separate for 2010, 2011 and 2012
df_accidents_vehicles_2010_2011 = pd.merge(df_accidents_vehicles_2010, df_accidents_vehicles_2011, on='Districts', suffixes=('_2010', '_2011'))
df_accidents_vehicles_2010_2011.head(30)
#add _2012 suffix to the columns of df_accidents_vehicles_2012
df_accidents_vehicles_2012.columns = [str(col) + '_2012' for col in df_accidents_vehicles_2012.columns]
df_accidents_vehicles_2012.head(30)
#rename the Districts_2012 column to Districts
df_accidents_vehicles_2012.rename(columns={'Districts_2012':'Districts'}, inplace=True)
df_accidents_vehicles_2010_2011_2012 = pd.merge(df_accidents_vehicles_2010_2011, df_accidents_vehicles_2012, on='Districts') 
df_accidents_vehicles_2010_2011_2012.head(30)
#create new df with the total columns for 2010, 2011 and 2012 and Districts
df_accidents_vehicles_2010_2011_2012_total = df_accidents_vehicles_2010_2011_2012[['Districts', 'Total_2010', 'Total_2011', 'Total_2012']]
df_accidents_vehicles_2010_2011_2012_total.rename(columns={'Total_2010':'Total_accidents_2010', 'Total_2011':'Total_accidents_2011', 'Total_2012':'Total_accidents_2012'}, inplace=True)
df_accidents_vehicles_2010_2011_2012_total.head(30)
```

```python
#concatenate the df_accidents_injuries_2010_2011_2012_total and df_accidents_vehicles_2010_2011_2012_total data frames on Districts
df_accidents_2010_2011_2012 = pd.concat([df_accidents_injuries_2010_2011_2012_total, df_accidents_vehicles_2010_2011_2012_total], axis=1)
#drop the duplicate Districts column
df_accidents_2010_2011_2012 = df_accidents_2010_2011_2012.loc[:,~df_accidents_2010_2011_2012.columns.duplicated()]
df_accidents_2010_2011_2012.head(30)
```

```python
import matplotlib.pyplot as plt

import numpy as np

# Assuming df_accidents_2010_2011_2012 has columns for each year
years = ['Total_accidents_2010', 'Total_accidents_2011', 'Total_accidents_2012']
years2 = ['Total_injuries_2010', 'Total_injuries_2011', 'Total_injuries_2012']
x_years = ['2010', '2011', '2012']

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Width of a bar 
width = 0.3       

# Plotting
for i, (year, year2) in enumerate(zip(years, years2)):
    ax.bar(i - width/2, df_accidents_2010_2011_2012[year], width, label='Accidents', color='skyblue')
    ax.bar(i + width/2, df_accidents_2010_2011_2012[year2], width, label='Injuries', color='thistle')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Year')
ax.set_ylabel('Counts')
ax.set_title('Accidents vs Injuries per year')
ax.set_xticks(np.arange(len(years)))
ax.set_xticklabels(x_years)

ax.set_facecolor('seashell')

# Show the plot
plt.show()
#save the plot
fig.savefig('accidents_vs_injuries.png')
```
